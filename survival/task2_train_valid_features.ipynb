```python
cols = ['player 0', 'score 0', 'player 1', 'score 1']

for col in cols:
    survival_train [col] =survival_train[col].apply(lambda x: np.float64(x) if pd.notnull(x) else x)
```

```python
survival_train.head()
```

```python
player0_counts = survival_train['player 0'].value_counts()
player0_counts
```

```python
play_counts = {}
unique_dates = {}

for column in ['player 0', 'player 1', 'player 2', 'player 3', 'player 4', 'player 5', 'player 6', 'player 7']:
    valid_rows = survival_train[survival_train[column].notna()]
    
    for player in valid_rows[column].unique():
        play_counts[player] = play_counts.get(player, 0) + valid_rows[valid_rows[column] == player].shape[0]

        if player not in unique_dates:
            unique_dates[player] = set()
        unique_dates[player].update(valid_rows[valid_rows[column] == player]['date'].unique())

play_perday = {player: total / len(unique_dates[player]) for player, total in play_counts.items()}
```

```python
print(play_counts)
print(len(play_counts))
```

```python
print(play_perday)
print(len(play_perday))
```

```python
total_scores = {}
unique_dates = {}

for i in range(8):
    player_col = f'player {i}'
    score_col = f'score {i}'

    valid_rows = survival_train[survival_train[player_col].notna()]

    for index, row in valid_rows.iterrows():
        player = row[player_col]
        score = row[score_col]

        total_scores[player] = total_scores.get(player, 0) + score

        if player not in unique_dates:
            unique_dates[player] = set()
        unique_dates[player].add(row['date'])


scores_perday = {player: total_scores[player] / len(unique_dates[player]) for player in total_scores}
```

```python
print(total_scores)
print(len(total_scores))
```

```python
print(scores_perday)
print(len(scores_perday))
```

```python
scores_pergame = {key: total_scores[key] / play_counts[key] for key in total_scores if key in play_counts}

print(scores_pergame)
print(len(scores_pergame))
```

```python
player_rankings = {}

for index, row in survival_train.iterrows():
    scores = {row[f'player {i}']: row[f'score {i}'] for i in range(8) if pd.notna(row[f'player {i}'])}

    rankings = pd.Series(scores).rank(ascending=False)

    for player_index, rank in rankings.items():
        player_rankings[player_index] = player_rankings.get(player_index, 0) + rank

print(player_rankings)
print(len(player_rankings))
```

```python
ranking_pergame = {key: player_rankings[key] / play_counts[key] for key in player_rankings if key in play_counts}

print(ranking_pergame)
print(len(ranking_pergame))
```

```python
features2 = pd.DataFrame({'play_counts': play_counts, 'play_perday': play_perday, 'total_scores': total_scores, 'scores_perday': scores_perday, 'scores_pergame': scores_pergame, 'ranking_pergame': ranking_pergame})
features2
```

```python
features2 = features2.sort_index()
features2
```
